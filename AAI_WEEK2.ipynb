```python
import numpy as np

def step_function(x):
    return 1 if x >= 0 else 0
def train_perceptron(X, T, lr):
    w = np.zeros(X.shape[1])
    b = 0
    epoch = 0

    print("\nInitial Weights:", w)
    print("Initial Bias:", b)

    while True:
        epoch += 1
        total_error = 0
        print(f"\nEpoch {epoch}")

        for i in range(len(X)):
            yin = np.dot(X[i], w) + b
            y = step_function(yin)

            error = T[i] - y
            total_error += abs(error)


            w = w + lr * error * X[i]
            b = b + lr * error

            print(f"Input: {X[i]}, Target: {T[i]}, Output: {y}")
            print(f"Weights: {w}, Bias: {b}")

        if total_error == 0:
            print("\nTraining completed successfully!")
            break

    return w, b
```

```python

n = int(input("Enter number of inputs (2 for AND/OR gate): "))

p = int(input("Enter number of input patterns: "))

X = []
T = []

print("\nEnter input patterns:")
for i in range(p):
    inputs = list(map(int, input(f"Pattern {i+1} inputs: ").split()))
    X.append(inputs)

print("\nEnter target outputs:")
for i in range(p):
    target = int(input(f"Target for pattern {i+1}: "))
    T.append(target)

X = np.array(X)
T = np.array(T)

lr = float(input("\nEnter learning rate: "))

w, b = train_perceptron(X, T, lr)

print("\nFinal Weights:", w)
print("Final Bias:", b)
```
